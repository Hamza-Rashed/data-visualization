## Health (Life Expectancy)

```python

first_20=df['Country'].head(20)
f20=df['Health (Life Expectancy)'].head(20)
```

```python
Country_li=[]
for i in first_20:
    Country_li.append(str(i))
Economy_li2=[]
for i in f20:
    Economy_li2.append(float(i))

df = pd.DataFrame(Economy_li2, index=Country_li, columns=['x'])
 
df.plot(kind='pie', subplots=True, figsize=(8, 8))
```
